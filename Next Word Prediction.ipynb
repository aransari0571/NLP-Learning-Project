```python
df = pd.read_csv('tmdb_5000_movies.csv')
df.head()
```

```python
df = df['original_title']
```

```python
df
```

```python
movie_name = df.to_list()
```

```python
movie_name
```

```python
tokenizer = tf.keras.preprocessing.text.Tokenizer()

tokenizer.fit_on_texts(movie_name)


seq = tokenizer.texts_to_sequences(movie_name)
```

```python
seq[:10]
```

```python
tokenizer.word_index
```

```python
x = []
y = []
total_words_dropped = 0

for i in seq:
    if len(i) > 1:
        for index in range(1, len(i)):
            x.append(i[:index])
            y.append(i[index])
    else:
        total_words_dropped += 1

print("Total Single Words Dropped are:", total_words_dropped)
```

```python
X = tf.keras.preprocessing.sequence.pad_sequences(x)
```

```python
X
```

```python
X.shape
```

```python
y = tf.keras.utils.to_categorical(y)
```

```python
y
```

```python
y.shape
```

```python
vocab_size = len(tokenizer.word_index) + 1
```

```python
vocab_size
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(vocab_size, 14),
    tf.keras.layers.LSTM(100, return_sequences=True),
    tf.keras.layers.LSTM(100),
    tf.keras.layers.Dense(100, activation="relu"),
    tf.keras.layers.Dense(vocab_size, activation="softmax"),
])
```

```python
model.compile(
         optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
         loss="categorical_crossentropy",
         metrics=["accuracy"]
     )
```

```python
model.fit(X,y, epochs=50)
```

```python
model.save('nwp.h5')
```

```python
vocab_array = np.array(list(tokenizer.word_index.keys()))
```

```python
vocab_array
```

```python
def make_prediction(text, n_words):
    for i in range(n_words):
      text_tokenize = tokenizer.texts_to_sequences([text])
      text_padded = tf.keras.preprocessing.sequence.pad_sequences(text_tokenize, maxlen=14)
      prediction = np.squeeze(np.argmax(model.predict(text_padded), axis=-1))
      prediction = str(vocab_array[prediction - 1])
      print(vocab_array[np.argsort(model.predict(text_padded)) -1] .ravel()[:-3])
      text += " " + prediction
    return text
```

```python
make_prediction("spider", 10)
```

```python
make_prediction("cloudy", 10)
```

```python
make_prediction("thor", 10)
```

```python
import gradio as gr
```

```python
interface = gr.Interface(
    fn=make_prediction,
    inputs=[
        gr.Textbox(lines=2, placeholder="Enter starting text here..."),
        gr.Slider(minimum=1, maximum=15, step=1, label="Number of words to predict")
    ],
    outputs=gr.Textbox(label="Predicted Text"),
    title="Next Word Prediction (Movie Titles)",
    description="Enter a starting word/phrase and get predictions for movie titles."
)

interface.launch()
```
